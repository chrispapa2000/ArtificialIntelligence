```python
def histogram(df, class_feature):
    """
    Η συνάρτηση αυτή δέχεται σαν ορίσματα τα ονόματα των στηλών που περιέχουν αριθμητικές τιμές και το όνομα της στήλης που περιέχει την ετικέτα 
    και τυπώνει τα ιστογράμματα σε σχέση με την ετικέτα τους
    Είσοδος:
        df: pd.DataFrame
        class_feature: το όνομα της στήλης του df που περιέχει την ετικέτα 
    """
    if df is not None: # αν η είσοδος δεν είναι κένη
        plt.figure(figsize = [40, 40]) # ορίζουμε ένα διάγραμμα
        for i, f in enumerate(df.columns): # για κάθε μια στήλη της εισόδου
            plt.subplot(7, 2, i + 1) # φτιάχνουμε ένα subplot (υποδιάγραμμα του συνολικού)
            g = sns.histplot(data = df, x = df[f], hue = df[class_feature], multiple  = 'dodge', kde=True) # φτιάχνουμε το ιστόγραμμα
            g.set_xticklabels(g.get_xticklabels(), rotation=90)
            g.xaxis.get_label().set_fontsize(20)
            g.yaxis.get_label().set_fontsize(20)
        plt.plot() # και το τυπώνουμε 


## Συμπληρώστε τα πεδία τα όποια θέλετε να εξετάσετε 
hist_features = ["title_year", "budget", "vote_average", "popularity", "gross", ] # ποία άλλα πεδία θέλετε να εξετάσετε μέσω του histogram;
histogram(dataset[hist_features + ["cat_ratings"]], class_feature="cat_ratings")
```

```python
from itertools import chain
# preprocess values 
def MultiLabelEncoder(dataset: pd.DataFrame, column_name: str, return_names=False):
    """

    Η συνάρτηση αυτή μετατρέπει μια στήλη με πολλαπλές τιμές σε κάθε κελί σε ένα δίανυσμα. Τέτοιες στήλες στο σύνολό μας είναι
    η στήλη  'genres' και 'plot_keywords'.
    Αν για παράδειγμα το genres για μια ταινία είναι "Action"|"Horror" ενώ τα συνολικά genres είναι k
    τότε η συνάρτηση πρέπει να επιστρέφει ένα διάνυσμα k θέσεων με μηδενικά εκτός από τις θέσεις που αντιστοιχούν στο Action και το Horror,
    όπου θα είναι 1. 
    Είσοδος:
        dataset: ένα dataframe
        column_name: το όνομα του κελίου το όποιο θέλουμε να μετατρέψουμε σε διάνυσμα

    'Έξοδος: ένα διάνυσμα 
    """

    ##################
    ## Your code below

    df = dataset[column_name]
    df = df.apply(lambda x: x.split('|') if isinstance(x,str) else [])

    # extract unique values
    values = list(df.values)
    values = set(chain.from_iterable(values))

    res = list(df)
    res = map(lambda items : [(1 if val in items else 0) for val in values] , res)
    
    ## Your code above
    ##################
    if return_names:
        return list(res), values
    return list(res) # το δίανυσμα που κωδικοποιεί την τιμή του κελίου
```

```python
import numpy as np

class Node:
    def __init__(self, feature_index=None, threshold=None, value=None, left=None, right=None):
        self.feature_index = feature_index  # Index of the feature to split on
        self.threshold = threshold  # Threshold value for the feature
        self.value = value  # Predicted value at the node (for leaf nodes)
        self.left = left  # Left child node
        self.right = right  # Right child node

class CARTClassifier:
    def __init__(self, max_depth=None, min_samples_split=2):
        self.max_depth = max_depth  # Maximum depth of the tree
        self.min_samples_split = min_samples_split  # Minimum number of samples required to split a node

    def fit(self, X, y):
        self.num_classes = len(np.unique(y))
        self.tree = self._build_tree(X, y)

    def predict(self, X):
        return [self._predict_sample(x, self.tree) for x in X]

    def _build_tree(self, X:list, y:list, depth=0):
        num_samples, num_features = X.shape
        ##################
        ## Your code below
        gini = self._gini_index(y)
        num_samples_per_class = np.count_nonzero(y==0), np.count_nonzero(y==1)
        ## Your code above
        ##################

        if depth == self.max_depth or num_samples < self.min_samples_split or gini == 0:
            leaf_value = np.argmax(num_samples_per_class)
            return Node(value=leaf_value)

        best_gini = float('inf')
        best_feature_index = None
        best_threshold = None

        for feature_index in range(num_features):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                left_indices = X[:, feature_index] <= threshold
                right_indices = X[:, feature_index] > threshold                

                left_gini = self._gini_index(y[left_indices])
                right_gini = self._gini_index(y[right_indices])
               
                weighted_gini = (np.count_nonzero(left_indices==True)/num_samples)*left_gini + (np.count_nonzero(right_indices==True)/num_samples)*right_gini
                if weighted_gini < best_gini:
                    best_gini = weighted_gini
                    best_feature_index = feature_index
                    best_threshold = threshold

        left_indices = X[:, best_feature_index] <= best_threshold
        right_indices = X[:, best_feature_index] > best_threshold
        # print(left_indices, right_indices)

        left_child = self._build_tree(X[left_indices], y[left_indices], depth + 1)
        right_child = self._build_tree(X[right_indices], y[right_indices], depth + 1)

        return Node(feature_index=best_feature_index, threshold=best_threshold, left=left_child, right=right_child)

    def _gini_index(self, y:np.array):
        ##################
        ## Your code below
        if len(y)==0:
            # print("length = 0")
            return 0
        pos = np.count_nonzero(y==1) / len(y)
        neg = np.count_nonzero(y==0) / len(y)
        gini = 1 - (pos**2) - (neg**2)
        ## Your code above
        ##################
        return gini # return a single value

    def _predict_sample(self, x, node):
        if node.value is not None:
            return node.value

        if x[node.feature_index] <= node.threshold:
            return self._predict_sample(x, node.left)
        else:
            return self._predict_sample(x, node.right)
```

```python
def plot_metrics(dicts_to_plot, labels, colors, title, yticks, pos=111):
    plt.subplot(pos)
    for d,c in zip(dicts_to_plot,colors):
        plt.plot(list(d.keys()), list(d.values()), color=c)

    plt.title(title)
    plt.yticks(yticks)
    plt.grid()
    plt.legend(labels)
    
    plt.xlabel('Max Depth')
    plt.ylabel('Metric(%)')
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

test_pr = {}
test_recall = {}
test_f1 = {}

for depth in [2, 3, 4, 5, 6, 7, 8, 9, 10, 15, 20, 25, 30]:
    dt_model = DecisionTreeClassifier(max_depth = depth) 
    dt_model.fit(X_train,Y_train) 

    y_pred_test = dt_model.predict(X_test)

    test_pr[depth] = precision_score(Y_test, y_pred_test)*100
    test_recall[depth] = recall_score(Y_test, y_pred_test)*100
    test_f1[depth] = f1_score(Y_test, y_pred_test)*100
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

test_pr_forest = {}
test_recall_forest = {}
test_f1_forest = {}

for depth in [2, 3, 4, 5, 6, 7, 8, 9, 10, 15, 20, 25, 30]:
    dt_model = RandomForestClassifier(n_estimators=50, max_depth = depth) 
    dt_model.fit(X_train,Y_train) 

    y_pred_test = dt_model.predict(X_test)

    test_pr_forest[depth] = precision_score(Y_test, y_pred_test)*100
    test_recall_forest[depth] = recall_score(Y_test, y_pred_test)*100
    test_f1_forest[depth] = f1_score(Y_test, y_pred_test)*100
```

```python
plt.figure(figsize=(20,5))
plot_metrics(dicts_to_plot=[test_pr, test_recall, test_f1], labels=['precision', 'recall', 'f1_score'], colors=['cornflowerblue', 'green', 'orange'],
             title='Decision Tree Results', yticks=[i for i in range(30,100,5)],pos=131)

plot_metrics(dicts_to_plot=[test_pr_forest, test_recall_forest, test_f1_forest], labels=['precision', 'recall', 'f1_score'], colors=['cornflowerblue', 'green', 'orange'],
             title="Random Forest Results (n_enstimators=50)", yticks=[i for i in range(30,100,5)],pos=132)

plot_metrics(dicts_to_plot=[test_f1, test_f1_forest], labels=['decision tree', 'random forest'], colors=['goldenrod', 'firebrick'], title='Common plot for f1-scores',
             yticks=[i for i in range(30,100,5)], pos=133)
```

Ωστόσο, ιδιαίτερα στο κοινό διάγραμμα για τα f1-scores φαίνεται ότι το decision tree πετυχαίνει την καλύτερη τιμή του για σημαντικά μικρότερο βάθος (μεταξύ 5 και 10) σε σχέση με το random forest που βελτιστοποιείται για βάθος ίσο με 30. Δεδομένου ότι το decision tree αποτελεί απλούστερο αλγόριθμο, αλλά και το μικρότερο βάθος δίνει πιο απλές και διαισθητικές λύσεις, μπορούμε να πούμε ότι το Decision Tree είναι καλύτερη λύση για το πρόβλημά μας σε σχέση με το random forest.
